## Mathematical Model

We are treating the droplet of water as a damped harmonic oscillator, first proposed by R.S.Shaw in ”The Dripping Faucet as a Model Chaotic System”. Using Newton's laws of motions, we get the following differential equation for a damped harmonic oscillator. In our case, the mass of oscillator is not constant, but is changing at a constant rate as the water flows to the droplet. 

$ \frac{d(mv)}{dt}= mg - kx - bv $  which expands to $\frac{d^2 x}{dt^2} = mg - kx - bv - \frac{dm}{dt}v$

$\frac{dm}{dt} = r $

$\frac{dx}{dt} = v$

Constants  
$m$ : mass of the water  
$g$ : acceleration due to gravity  
$k$ : the spring constant  
  $b$ : the damping constant    
  $r$ : the rate of water flow (mass per unit time)

Additionaly, this model has two more constraints which give it the leaky faucet behavior. As the $x$ value approaches and obtains a particular critical value $x_0$, the water droplet detaches itself from the bulk of the system. The mass of the water droplet that leaves $\Delta m$ is proportional to the velocity at the time of detachment. Further, considering the intertia of the water drop, we want the mass of drop to be smaller when velocity is low, and for high velocities, the mass of the droplet should be closer to its limiting value $m$.

Constraints:  
$ x_0 = constant$  
$ \Delta m = \frac{v}{\beta+ v}$ where $v = \frac{dx}{dt}$

Initial Conditions:  
$x(t=0) = 0$ and $v(t=0) = 0$

When $x=x_0$ at some time $t_0$, we reset values to following:  
$x(t_0) = 0$ and $v(t_0) = 0$  
$m$ = $m - \Delta(m)$

The differential equation for the model will be solved numerically using a modified Runge Kutta method which takes into account the additional constraints. 

## Differential Equation Solver

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#derivative functions

def f1(x,v):
    return v

def f2(m,g,k,b,r,x,v):
    return (-k/m*x - b/m*v + g - v*r/m)
```

```python
def runge_kutta(m_i,g,k,b,r,beta,N=3000,h=0.01,xc=3):
    x = np.zeros(N+1)
    v = np.zeros(N+1)
    t = np.zeros(N+1)
    m = np.zeros(N+1)
    T = []
    end_mass= []
    x[0]=0
    v[0]=0
    t[0] =0
    m[0] = m_i
    las_t = 0
    for i in range(1,N+1):
        m[i] = m[i-1]+h*r
        t[i] = t[i-1]+h
        kx1 = f1(x[i-1],v[i-1])
        kv1 = f2(m[i-1],g,k,b,r,x[i-1],v[i-1])
        kx2 = f1(x[i-1]+h*kx1/2, v[i-1]+h*kv1/2)
        kv2 = f2(m[i-1],g,k,b,r,x[i-1]+h*kx1/2, v[i-1]+h*kv1/2)
        kx3 = f1(x[i-1]+h*kx2/2, v[i-1]+h*kv2/2)
        kv3 = f2(m[i-1],g,k,b,r,x[i-1]+h*kx2/2, v[i-1]+h*kv2/2)
        kx4 = f1(x[i-1]+h*kx3, v[i-1]+h*kv3)
        kv4 = f2(m[i-1],g,k,b,r,x[i-1]+h*kx3, v[i-1]+h*kv3)
        x[i] = x[i-1]+(kx1+2*kx2+2*kx3+kx4)*h/6
        v[i] = v[i-1]+(kv1+2*kv2+2*kv3+kv4)*h/6
        if(x[i]>=xc):
            x[i] = x[0]
            end_mass.append(m[i])
            m[i] = m[i] - (v[i]/(beta+v[i]))*m[i]
            v[i] = 0
            dummy = t[i] - las_t
            T.append(dummy)
            las_t = t[i]
    return x,v,t,m,T,end_mass
    
```

## Periodic Behavior

```python
x,v,t,m,T,end_mass = runge_kutta(m_i=0.000001,g=9.8,k =0.01,b=0.0003,r=0.000037,beta=0.0286,N=80000,h=0.001,xc=0.05)
```

```python
fig,ax = plt.subplots(figsize=(12, 3), ncols=3, nrows =1)
ax[0].plot(t,-x)
ax[1].plot(-x[20000:],v[20000:], color ='lightcoral') #Printing 
ax[0].set_xlim(0,10)
#fig.set_title("Period One Behavior")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Position")

ax[1].set_xlabel("Position")
ax[1].set_ylabel("Velocity")
#plt.show()

for i in range(5,len(T)-1):
        ax[2].plot(T[i],T[i+1],'.')
ax[2].set_xlim(0.5,1)
ax[2].set_ylim(0.5,1)

ax[2].set_xlabel("T[i]")
ax[2].set_ylabel("T[i+1]")
fig.tight_layout()
ax[0].set_title("Evolution of system with time")
ax[1].set_title("Phase diagram")
ax[2].set_title("Attractor")
plt.savefig('SimplePeriodic.eps', format='eps', dpi=1000)
```
